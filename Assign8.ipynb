# Offline Text-to-Speech (pyttsx3)

This example uses the local `pyttsx3` library (no API key). It saves `output.wav` in the current folder.

```python
import sys

# Ensure pyttsx3 is installed
try:
    import pyttsx3
except ImportError:
    print("Installing pyttsx3...")
    %pip install pyttsx3
    import pyttsx3
```

```python
import pyttsx3

text = "Hello! This is a short offline text to speech demo using pyttsx3."
output_path = "output.wav"

engine = pyttsx3.init()
engine.save_to_file(text, output_path)
engine.runAndWait()

output_path
```
